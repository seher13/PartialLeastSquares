```python
from math import sqrt
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import matplotlib.pyplot as plt
```

```python
water_plot = []
fat_plot = []
protein_plot = []
```

```python
for n_comp in range(1,101):
  my_plsr = PLSRegression(n_components=n_comp, scale=True)
  my_plsr.fit(x_train, y_train)
  preds = my_plsr.predict(x_val)

  water_rmse = sqrt(mean_squared_error(y_val[:,0], preds[:,0]))
  fat_rmse = sqrt(mean_squared_error(y_val[:,1], preds[:,1]))
  protein_rmse = sqrt(mean_squared_error(y_val[:,2], preds[:,2]))

  water_plot.append( water_rmse)
  fat_plot.append ( fat_rmse)
  protein_plot.append(protein_rmse)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(18,6))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.plot(range(1, 101), water_plot)
ax2.plot(range(1,101), fat_plot)
ax3.plot(range(1,101), protein_plot)

ax1.set_title('Water prediction RMSE vs. components no')
ax1.set_xlabel('Components no')
ax1.set_ylabel('RMSE')

ax2.set_title('Fat prediction RMSE vs. components no')
ax2.set_xlabel('Components no')
ax2.set_ylabel('RMSE')

ax3.set_title('Protein prediction RMSE vs. components no')
ax3.set_xlabel('Components no')
ax3.set_ylabel('RMSE')
```

```python
from sklearn.metrics import r2_score

my_plsr_test = PLSRegression(n_components=15, scale=True)
my_plsr_test.fit(x_train, y_train)
preds_test = my_plsr_test.predict(x_val)
```

```python
r2_test = r2_score(y_val, preds_test)
r2_test
```

```python
# find best value for n_comp
from sklearn.metrics import r2_score

best_r2 = 0
best_ncomp = 0
for n_comp in range(1,101):
  my_plsr = PLSRegression(n_components=n_comp, scale=True)
  my_plsr.fit(x_train, y_train)
  preds = my_plsr.predict(x_val)

  r2 = r2_score(y_val, preds)
  if r2 > best_r2:
    best_r2 = r2
    best_ncomp = n_comp
  
print(best_r2, best_ncomp)
```

```python
preds.shape
```

```python
my_plsr_best = PLSRegression(n_components=best_ncomp, scale=True)
my_plsr_best.fit(x_train, y_train)
preds_best = my_plsr_best.predict(x_test)
r2_score(y_test, preds_best)
```

Conclusion: We have successfuly crated PLS model which predicts meat composition (water, fat, protein contents), from chemometric measurements. We have tested for optimum components number in PLS analysis, using train and validation data set in order to avoid overfitting. Optimum number of components was 15 and R^2 on validation data set was 0.94. Therefore, we created a model with 15 components, which we used to calculate R^2 on test set, to make sure we did not overfit. Result was very good, with R^2=0.96.

#Example 2
## Predict olive oil country of origin from chemical adn sensory measurements (optimize explainability over prediction)

```python
data_oo = pd.read_csv('s3://articledatas3/olive_oil.csv')
data_oo = data_oo.drop('Unnamed: 0', axis=1)
```

```python
data_oo.head()
```

```python
# get dummies will dummy code your categorial variables (also called one-hot encoding)

dummified_data = pd.get_dummies(data_oo)
dummified_data
```

```python
data_oo
```

```python
y = dummified_data.iloc[:,-3:]
```

```python
y
```

```python
x = dummified_data.iloc[:,:-3]
```

```python
x
```

```python
from sklearn.cross_decomposition import PLSRegression
```

```python
my_plsr = PLSRegression(n_components=2, scale=False)
my_plsr.fit(x, y)
```

```python
# hand made biplot in Python

# extract scores (one score per individual per component)

scores_df = pd.DataFrame(my_plsr.x_scores_)
```

```python
scores_df
```

```python
std_scores_dim1 = 2*((scores_df[0] - min(scores_df[0]))/(max(scores_df[0]) - min(scores_df[0]))) - 1
std_scores_dim2 = 2*((scores_df[1] - min(scores_df[1]))/(max(scores_df[1]) - min(scores_df[1]))) - 1
```

```python
std_scores_dim1
```

```python
# extract loadings (one loading per variable per component)

loadings_df = pd.DataFrame(my_plsr.x_loadings_)
```

```python
loadings_df
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
plt.xlim((-1.2,1.2))
plt.ylim((-1.2,1.2))

for i, var_i in enumerate(dummified_data.columns[0:11]):
  x_load = loadings_df.iloc[i,0]
  y_load = loadings_df.iloc[i,1]

  ax.arrow(0,0,x_load, y_load, color='red')
  ax.text(x_load, y_load, var_i, color='red')

for i in range(len(std_scores_dim1)):
  x_score = std_scores_dim1[i]
  y_score = std_scores_dim2[i]

  ax.arrow(0,0,x_score,y_score, color='blue')
  ax.text(x_score,y_score, data_oo.loc[i,'countries'], color='blue')

ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_title('Score and loading plot for Components 1 and 2')

```

Conclusion: We can see that we succeded in approximate classification of olive oil origins between Spain and Greece, however we failed to isolate Italy completly from Spain anf Greece. Spain and Greece are divided by component 1. We can see that component one is mostly influenced by yellow and green color. Therefore we can see that Greek olive oil have more green appearance, whereas Spanish olive oils have more yellowish appearance. Second component is the most influenced by brown color on one side and transparent ang glossy appearance on the other side. There is slight split between olive oils from Italy, which are in general more brown and less glossy and transparent as olive oils from Greence and Spain.
